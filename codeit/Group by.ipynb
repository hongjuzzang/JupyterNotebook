```python
import pandas as pd

df = pd.read_csv('data/laptops.csv')
df
```

```python
brand_nation={
    'Dell':'U.S.',
    'Apple': 'U.S.',
    'Acer':'Taiwan',
    'HP':'U.S.',
    'Lenovo':'China',
    'Alienware':'U.S.',
    'Microsoft':'U.S.',
    'Asus':'Taiwan'
}
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
nation_groups.plot(kind='box',y='price')
```

```python
nation_groups.plot(kind='hist',y='price')
```
