## 강의실 배정하기 1

```python
df = pd.read_csv('data/enrolment_2.csv')
df
```

```python
counts = df['course name'].value_counts()
```

```python
df
```

```python
audi = list(counts[(counts>=80)].index)
large = list(counts[(counts<80) & (counts>=40)].index)
medium = list(counts[(counts<40) & (counts>=15)].index)
small = list(counts[(counts<15) & (counts>=5)].index)
```

```python
large
```

```python
for course in audi:
    df.loc[df['course name']==course,'room assignment']="Auditorium"
for course in large:
    df.loc[df['course name']==course,'room assignment']="Large room"
for course in medium:
    df.loc[df['course name']==course,'room assignment']="Medium room"
for course in small:
    df.loc[df['course name']==course,'room assignment']="Small room"
```

```python
df.loc[df['status']=='not allowed','room assignment'] = 'not assigned'
```

```python
df.head(10)
```

### 강의실 배정하기2

```python
df = pd.read_csv('data/enrolment_3.csv')
df
```

```python
df.rename(columns = {'room assignment':'room number'},inplace=True)
```

```python
df.loc[df['status']=='not allowed','room number']='not assigned'
```

```python
counts = df['room number'].value_counts()
counts
```

```python
audi = sorted(list(df.loc[df['room number']=='Auditorium','course name'].value_counts().index))
large = sorted(list(df.loc[df['room number']=='Large room','course name'].value_counts().index))
medium = sorted(list(df.loc[df['room number']=='Medium room','course name'].value_counts().index))
small = sorted(list(df.loc[df['room number']=='Small room','course name'].value_counts().index))
```

```python
large
```

```python
for idx,course in zip(range(1,len(audi)+1),audi):
    df.loc[df['course name']==course,'room number']="Auditorium-"+str(idx)

for idx,course in zip(range(1,len(large)+1), large):
    df.loc[df['course name']==course,'room number']="Large-"+str(idx)

for idx,course in zip(range(1,len(medium)+1), medium):
    df.loc[df['course name']==course,'room number']="Medium-"+str(idx)

for idx,course in zip(range(1,len(small)+1), small):
    df.loc[df['course name']==course,'room number']="Small-"+str(idx)
```
